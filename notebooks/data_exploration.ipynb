```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

num_samples=16

images = [os.path.join('data/raw/train', f'{img}.png') for img in range(1, num_samples+1)]
sample_images = images[:num_samples]

plt.figure(figsize=(10, 10))
for i, img_path in enumerate(sample_images):
    img = mpimg.imread(img_path)
    plt.subplot(4, 4, i+1)
    plt.imshow(img)
    plt.axis('off')
plt.tight_layout()
plt.savefig('results/plots/sample_images.png')
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='label', order=data['label'].value_counts().index)
plt.title('Class Distributions in CIFAR-10 Dataset')
plt.xticks(rotation=45)
plt.savefig('results/plots/class_distributions.png')
plt.show()
```
